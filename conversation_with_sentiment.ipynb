```python
import pandas as pd
from transformers import pipeline
```

```python
# Load sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis") #data2822029
```

```python
# Load your Excel file
df = pd.read_csv("data2822029.csv")
```

```python
# Function to extract only Human lines from the conversation
def extract_human_lines(convo):
    return "\n".join([line.split("Human: ")[1] for line in convo.split("\n") if "Human:" in line])
```

```python
# Apply to dataframe
df['Human_Sayings'] = df['Conversation'].apply(extract_human_lines)
```

```python
%%time
# Analyze sentiment
df['Sentiment'] = df['Human_Sayings'].apply(lambda x: sentiment_pipeline(x)[0]['label'] if x.strip() else 'NEUTRAL')
```

```python
print(df[['Human_Sayings', 'Sentiment']].head())
```

```python
# Save result
df.to_excel("conversation_with_sentiment.xlsx", index=False)
```
